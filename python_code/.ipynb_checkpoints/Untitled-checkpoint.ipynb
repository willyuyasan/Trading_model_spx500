```python
import requests
import bs4 as bs
import pickle

import datetime as dt
import os

import pandas as pd
import pandas_datareader.data as pdr

import yfinance as yf
```

```python
api_url = " https://demo-api-capital.backend-capital.com/"
response = requests.get(api_url)
response.json()
```

```python
#Data from yahoo 

api_url = 'http://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
html = requests.get(api_url)
soup = bs.BeautifulSoup(html.text, 'lxml')
table = soup.find('table', {'class': 'wikitable sortable'})
```

```python
tickers = []
for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text
        ticker = ticker[:-1]
        tickers.append(ticker)
```

```python
tickers
```

```python
with open("sp500tickers.pickle", "wb") as f:   
        pickle.dump(tickers, f)
```

```python
with open("sp500tickers.pickle", "rb") as f:
        tickers = pickle.load(f)
if not os.path.exists('stock_dfs'):
        os.makedirs('stock_dfs')
```

```python
start = dt.datetime(2022, 2, 25)
end = dt.datetime.now()
    
for ticker in tickers:
    if not os.path.exists('stock_dfs/{}.csv'.format(ticker)):
        df = pdr.DataReader(ticker.replace('.', '-'), 'yahoo', start, end)
        df.reset_index(inplace=True)
        df.set_index("Date", inplace=True)
        df.to_csv('stock_dfs/{}.csv'.format(ticker))
    else:
        print('Already have {}'.format(ticker))
```

```python
df = pdr.DataReader(['^GSPC'], 'yahoo', start, end, period="1m")
df.reset_index(inplace=True)
df.set_index("Date", inplace=True)
df.to_csv('stock_dfs/{}.csv'.format(ticker))
```
